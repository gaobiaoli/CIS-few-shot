```python
import matplotlib.pyplot as plt
import matplotlib
# print(matplotlib.matplotlib_fname())
matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams['font.size'] = 12
# 读取数据文件
alpha_values_CIS = []
acc_values_CIS = []

with open('/CV/gaobiaoli/project/github/CIS-few-shot/output/CIS/alpha.txt', 'r') as file:  # 假设文件名为 data.txt
    for line in file:
        parts = line.strip().split()
        alpha = float(parts[0].split(':')[1])
        acc = float(parts[2].split(':')[1])
        alpha_values_CIS.append(alpha)
        acc_values_CIS.append(acc)


alpha_values_mocs = []
acc_values_mocs = []

with open('/CV/gaobiaoli/project/github/CIS-few-shot/output/mocs/alpha.txt', 'r') as file:  # 假设文件名为 data.txt
    for line in file:
        parts = line.strip().split()
        alpha = float(parts[0].split(':')[1])
        acc = float(parts[2].split(':')[1])
        alpha_values_mocs.append(alpha)
        acc_values_mocs.append(acc)
plt.figure(figsize=(10, 6))
plt.plot(alpha_values_CIS, acc_values_CIS, linestyle='-', linewidth=2,label='CIS')
plt.plot(alpha_values_mocs, acc_values_mocs, linestyle='--', linewidth=2,label='MOCS')
plt.xlabel(r"$\alpha$", fontsize=16, weight='bold')
plt.ylabel("Accuracy (%)", fontsize=16, weight='bold')
# plt.title(r"Sensitivity Analysis of $\alpha$ on Accuracy", fontsize=18, weight='bold')
plt.grid(True, linestyle='--', alpha=0.7)

# 增加图例和调整边框
plt.legend(fontsize=15, loc='lower right', frameon=True, shadow=False, borderpad=1)
plt.savefig("sensitivity_analysis_alpha_acc.png", dpi=800, bbox_inches='tight')
```
